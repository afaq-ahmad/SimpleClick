```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from square_transfer import Square_Predictor

square_predictor = Square_Predictor(checkpoint='../drive/MyDrive/weights/cocolvis_vit_huge.pth')
```

```python
#uploading image to drive
image = cv2.cvtColor(cv2.imread('../audi2-1559742641.jpg'), cv2.COLOR_BGR2RGB)
image =  cv2.resize(image, (0, 0),fx=0.4, fy=0.4)
yolo_results = square_predictor.predict_rectangle(image)

masks = square_predictor.predict_mask_all(image,yolo_results,margin_per=0.10)
image = square_predictor.result_draw(image,yolo_results,masks)
```
