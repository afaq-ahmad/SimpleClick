```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from square_transfer import Square_Predictor

square_predictor = Square_Predictor(checkpoint='../drive/MyDrive/weights/cocolvis_vit_huge.pth')
```

```python
#uploading image to drive
image = cv2.cvtColor(cv2.imread('../audi2-1559742641.jpg'), cv2.COLOR_BGR2RGB)
image =  cv2.resize(image, (0, 0),fx=0.2, fy=0.2)
object_detection_mask = np.zeros((image.shape[0],image.shape[1]),'int32')
object_detection_mask[200:300,200:300]=1
pred_mask = square_predictor.predict_mask(image,object_detection_mask)
plt.imshow(image)
plt.show()
plt.imshow(object_detection_mask)
plt.show()
plt.imshow(pred_mask)
```
